```python
merged_data = pd.merge(data6, data4, on='id_student')
```

```python
merged_data.head()
```

```python
max_interaction = merged_data['sum_click'].max()
min_interaction = merged_data['sum_click'].min()
avg_interaction = merged_data['sum_click'].mean()
print(f"Hasil interaksi terbanyak adalah {max_interaction} kali.")
print(f"Hasil interaksi terendah adalah {avg_interaction} kali.")
print(f"Hasil interaksi terendah adalah {min_interaction} kali.")
```

```python
plt.bar(['Hasil Terbanyak', 'Hasil Sedang', 'Hasil Terendah'], [max_interaction, avg_interaction, min_interaction])
plt.xlabel('Waktu Interaksi')
plt.ylabel('Jumlah Mahasiswa Lulus')
plt.title('Hasil Interaksi Terbanyak dan Terendah')
plt.show()
```

```python
labels = ['Hasil Tertinggi', 'Hasil Sedang', 'Hasil Terendah']
sizes = [max_interaction, avg_interaction, min_interaction]
colors = ['#ff9999','#66b3ff','#99ff99']
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')
plt.title('Presentasi Hasil Interaksi Tertinggi, Sedang, dan Terendah')
plt.legend(labels, loc='upper left')
plt.show()
```
